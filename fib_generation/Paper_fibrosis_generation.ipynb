```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

```

```python
from matplotlib import colors

cmap = colors.LinearSegmentedColormap.from_list(
    'fibrosis', [(0, '#e2a858'),
                 (0.5, '#e2a858'),
                 (1, '#990102')])
```

```python
orig1 = plt.imread("original_texs/or_tex_25.png")
orig1 = rgb2gray(orig1)
orig1 = np.where(orig1 > 0.5, 1, 2)

orig2 = plt.imread("original_texs/or_tex_13.png")
orig2 = rgb2gray(orig2)
orig2 = np.where(orig2 > 0.5, 1, 2)

orig3 = plt.imread("original_texs/or_tex_34.png")
orig3 = rgb2gray(orig3)
orig3 = np.where(orig3 > 0.5, 1, 2)
```

```python
gen1 = np.load("paper_npy_original/generated_paper_1.npy")
gen2 = np.load("paper_npy_original/generated_paper_2.npy")
gen3 = np.load("paper_npy_original/generated_paper_3.npy")
```

```python
fig, ax = plt.subplots(3, 2, figsize=[7, 10])

for i in range(ax.shape[0]):
   for j in range(ax.shape[1]): 
        ax[i, j].get_xaxis().set_visible(False)
        ax[i, j].get_yaxis().set_visible(False)
        ax[i, j].patch.set_edgecolor('black')  
        ax[i, j].patch.set_linewidth(1)  

ax[0, 0].imshow(orig1, cmap=cmap)
ax[0, 1].imshow(gen1, cmap=cmap)

ax[1, 0].imshow(orig2, cmap=cmap)
ax[1, 1].imshow(gen2, cmap=cmap)

ax[2, 0].imshow(orig3, cmap=cmap)
ax[2, 1].imshow(gen3, cmap=cmap)

# fig.tight_layout()

plt.savefig("ds_generation_example.png")
```
